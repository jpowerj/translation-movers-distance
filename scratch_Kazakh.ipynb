### TMD Computation

```python
import glob
import os

import joblib

import trglobals
```

```python
trglobals.init_embedding_model()
```

```python
kk_fpaths = glob.glob("../Texts_Aligned/de_tge.kk_prog.pickled/*.pkl")
kk_fpaths
```

```python
all_dfs = []
for cur_pkl_fpath in kk_fpaths:
    cur_basename = os.path.basename(cur_pkl_fpath).replace(".pkl","")
    cur_df = pd.read_pickle(cur_pkl_fpath)
    cur_df.columns = ['de','ru']
    cur_df['chapter'] = cur_basename
    cur_df['tmd'] = trglobals.compute_tmd_df(cur_df)
    all_dfs.append(cur_df)
```

```python
full_df = pd.concat(all_dfs)
```

```python
full_df
```

```python
kk_cap_mean = trglobals.noninf_mean(full_df, 'tmd')
kk_cap_mean
```

```python
joblib.dump(kk_cap_mean, './lang_stats/cap_kk_mean.pkl')
```
